##### Import Library

```python
import numpy as np
import pandas as pd
```

##### Load Dataset

```python
titanic_data = pd.read_csv('train.csv')
titanic_data.head()
```

##### Memisahkan fitur - fitur / kolom yang penting dan yang tidak penting. Pada kasus ini, fitur yang tidak penting adalah PassengerId, Name, Ticket, Cabin  

```python
titanic_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
titanic_data.head()
```

### Memahami Data 
1. Survived coloum menjelaskan apakah penumpang selamat dari tenggelamnya kapal titanic atau tidak. <b>0 bernilai tidak selamat, sedangkan 1 bernilai selamat</b>

2. Pclass adalah kelas untuk seberapa banyak penumpang melakukan perjalanan, apakah pertama kali, kedua kalinya, dst

3. Sex adalah jenis kelamin. Pria atau Wanita

4. Age adalah umur penumpang kapal titanic

5. SibSp and Parch adalah jumlah saudara atau orang tua yang ikut atau berada dalam kapal titanic.

6. Embarked adalah menjelaskan di mana penumpang menaiki kapal. C = Cherbourg, Q = Queenstown, S = Southampton.

##### Mengubah kolom Gender agar memiliki nilai kategori dalam bentuk numerik berdasarkan Markdown memahami data. 

```python
from sklearn import preprocessing

le=preprocessing.LabelEncoder()
titanic_data['Sex'] = le.fit_transform(titanic_data['Sex'].astype(str))
titanic_data.head()
```

##### Kolom Embarked dilakukan proses One Hot Encoding, yaitu teknik untuk merepresentasikan sebuah data berbentuk categorical variable diubah ke dalam bentuk numerical value bernilai 0 dan 1

```python
titanic_data = pd.get_dummies(titanic_data,columns=['Embarked'])
titanic_data.head()
```

##### Melakukan pengecekan nilai data yang kosong / hilang

```python
titanic_data[titanic_data.isnull().any(axis=1)]
```

```python
#Drop data row yang kosong / hilang / tidak memiliki nilai
titanic_data = titanic_data.dropna()
```

##### Mean Shift Clustering

```python
#Import MeanShift dari library sklearn.cluster
from sklearn.cluster import MeanShift
#Mengatur nilai bandwith hyperparameter ke 30 agar kernel yang dihasilkan pendek dan lebar
analyzer = MeanShift(bandwidth=30) 
#Terapkan bandwith = 30 ke dalam dataset
analyzer.fit(titanic_data)
```

```python
#Fungsi pembantu untuk membantu estimasi nilai bandwith yang baik berdasarkan dataset
"""from sklearn.cluster import estimate_bandwith
estimate_bandwith(titanic_data)""" 
```

```python
labels = analyzer.labels_
```

```python
np.unique(labels)
```

##### Bandwith 30 menghasilkan 5 klaster. Setiap titik melakukan ke masing - masing klaster

```python
#Menambah kolom baru dalam dataset untuk menunjukan klaster data yang bergantungan
titanic_data['cluster_group'] = np.nan
data_length=len(titanic_data)
for i in range(data_length):
    titanic_data.iloc[i,titanic_data.columns.get_loc('cluster_group')] = labels[i]
titanic_data.head()
```

##### Melakukan EDA dengan Cluster

```python
#Kelompokan penumpang berdasarkan klaster nya
titanic_cluster_data = titanic_data.groupby(['cluster_group']).mean()
#Hitung penumpang untuk masing - masing klaster
titanic_cluster_data['Counts'] = pd.Series(titanic_data.groupby(['cluster_group']).size())
titanic_cluster_data
```
